```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('training.1600000.processed.noemoticon.csv', header=None, encoding="ISO-8859-1").sample(frac=0.1)
negative_text = ""
positive_text = ""

for row in df.iterrows():
    if row[1][0] == 0:
        negative_text += row[1][5]
    elif row[1][0] == 4:
        positive_text += row[1][5]

fig = plt.figure(figsize=(10, 10))
fig.add_subplot(2, 1, 1)
wordcloud = WordCloud(max_words=200).generate(negative_text)
plt.imshow(wordcloud, interpolation="bilinear")
plt.title("negative_text")
plt.axis("off")

# Adds a subplot at the 2nd position 
fig.add_subplot(2, 1, 2)
wordcloud = WordCloud(max_words=200).generate(positive_text)
plt.imshow(wordcloud, interpolation="bilinear")
plt.title("positive_text")
plt.axis("off")
```
